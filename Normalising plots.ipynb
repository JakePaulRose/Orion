```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import bin_tools
from wotan import flatten
import time
from OrionPY import OrionTools, Tools
from multiprocessing import Pool
```

```python
data = np.load('/data/jpr64/rebinned_data_test.npy', allow_pickle = True)
```

```python
lc = [x for x in data]

        
        
```

```python
def calculate_mag(fluxes):
    return -2.5 * np.log10(fluxes) + 20.2

def calc_mean_and_rms(fluxes):
        magnitudes = calculate_mag(fluxes)
        mean_mag = np.nanmean(magnitudes)
        rms_mag = np.sqrt(np.mean(np.square(mean_mag - magnitudes)))
        return mean_mag, rms_mag

def mag_propogated_errors(fluxes, flux_errors):
    return 1.48 * (2.5/np.log(10))* (flux_errors/fluxes)/np.sqrt(27)
    
```

```python
mean_mag_and_rms = [calc_mean_and_rms(x[:,1]) for x in lc]
x = np.array(mean_mag_and_rms)
mean_mags = x[:,0]
mean_rms = x[:,1]



```

```python
flux = [x[:,1] for x in lc]
flux_error = [x[:,2] for x in lc]

```

```python
flux_error_mag = np.array([np.nanmean(mag_propogated_errors(x,y)) for x,y in zip(flux, flux_error)])

```

```python
y = mean_rms/(flux_error_mag)
```

```python
%matplotlib notebook
plt.plot(mean_mags, y, linestyle = '', marker = 'o', markersize = 0.6)
plt.yscale('log')
plt.ylim(10**-2)
plt.xlabel(r'Mag')
plt.ylabel(r'Mean RMS / flux error (mag)')

```
